## Зaдание 1.2

### Напишите код на Python, реализующий расчет длины вектора, заданного его координатами.

```python
a = [10, 10, 10]
b = [0, 0, -10]
```

```python
def len_vector(vector):
    sum_x = 0
    for x in vector:
        sum_x += x**2
    len_vector = sqrt(sum_x)
    return len_vector
```

```python
len_a = len_vector(a)
len_a
```

```python
len_b = len_vector(b)
len_b
```

### 1) Решите систему уравнений:
### y = x2 – 1
### exp(x) + x∙(1 – y) = 1

```python
x = np.linspace(-3, 3, 100)
plt.plot(x, (np.exp(x) - 1 + x)/x)
plt.plot(x, x**2 - 1)
plt.xlabel('x')
plt.ylabel('y')
plt.ylim(-3, 10) 
plt.grid(True)
plt.show()

def eq(p):
    x, y = p
    return (y - x**2 + 1, np.exp(x) + x*(1 - y) - 1)
x1, y1 =  fsolve(eq, (2.5, 6))
print (x1, y1)
```
