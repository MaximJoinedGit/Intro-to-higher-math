## **<center>Элементы теории вероятностей</center>**

### 1. Найти вероятность выпадения 2 или 5 очков при подбрасывании игральной кости, на гранях которой имеются соответственно 1,2,3,4,5 и 6 очков.

Вероятность выпадения 2 или 5 равна $\frac{1}{6}$, т.к. число возможных событий 6, а число одно. Для того, чтобы найти вероятность в условии задачи, необходимо сложить вероятности, соответственно $\frac{1}{6} + \frac{1}{6} = \frac{1}{3}$

### 2. Найти вероятность того, что при двух подбрасываниях той же самой игральной кости сначала выпадет 2, а затем 5.

Вероятность выпадения каждой цифры, также как и в предыдущем задании, равна $\frac{1}{6}$, но теперь условие задачи сформировано иначе и у нас эти два события должны произойти друг за другом в четко определенной последовательности. Для того, чтобы найти вероятность этого события, нам нужно перемножить вероятности: $\frac{1}{6} \cdot \frac{1}{6} = \frac{1}{36}$

### 3. Найти вероятность выпадения 2 и 5 очков при двух подбрасываниях той же самой игральной игральной кости.

В этой задаче, необходимо найти вероятность выпадения комбинации 2-5 и 5-2. Вероятность 2-5 мы вычисляли в прошлом примере ($\frac{1}{36}$), а вероятность 5-2 такая же. Для того, чтобы найти вероятность, обозначенную в условии мы должны сложить две вероятности: $\frac{1}{36} + \frac{1}{36} = \frac{1}{18}$

### 4. На отрезке AB длиной 20 см наугад отметили точку C. Какова вероятность, что она находится на расстоянии не более 9 см от точки А и не более 15 см от точки В?

По сути у нас имеется отрезок от 5 до 9 включительно, и имеются предельно допустимые значения 0 и 20. Нам нужно найти вероятность, с которой точка будет находиться в этом отрезке: $\frac{4}{20} = \frac{1}{5}$

### 5. Телефонный номер состоит из 7 цифр. Какова вероятность, что это номер 8882227?

Телефонный номер не может начинаться с 0, поэтому первый номер будет начинаться с 1000000, а общее количество номеров будет $9 \cdot 10^6$. Вероятность, что какой-либо номер будет одним из этих равна: $\frac{1}{9000000}$

### 6. Набирая номер телефона, абонент забыл 2 последние цифры, и, помня только то, что эти цифры различны и среди них нет нуля, стал набирать их наудачу. Сколько вариантов ему надо перебрать, чтобы наверняка найти нужный номер? Какова вероятность того, что он угадает номер с первого раза?

Поскольку все цифры известны, кроме последних, а также мы знаем, что это не нули, то количество комбинаций будет равно $9 \cdot 9 = 81$, отнимем 9 комбинаций, в которых цифры одинаковы (11, 22, 33 и тд), получим 72 цифры нужно перепробовать, а вероятность, что он наберет ее с первого раза $\frac{1}{72}$.
