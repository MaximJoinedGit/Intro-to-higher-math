#### 1. Даны два вектора в трехмерном пространстве: (10,10,10) и (0,0,-10). Найдите их сумму. (на листочке) Напишите код на Python, реализующий расчет длины вектора, заданного его координатами. (в программе)


$\overrightarrow{A} = \begin{vmatrix} 10 \\ 10 \\ 10 \end{vmatrix}$
$\overrightarrow{B} = \begin{vmatrix} 0 \\ 0 \\ -10 \end{vmatrix}$

$\overrightarrow{C} = \overrightarrow{A} + \overrightarrow{B} = \begin{vmatrix} 10 \\ 10 \\ 10 \end{vmatrix} + \begin{vmatrix} 0 \\ 0 \\ -10 \end{vmatrix} = \begin{vmatrix} 10 + 0 \\ 10 + 0 \\ 10 - 10 \end{vmatrix} = \begin{vmatrix} 10 \\ 10 \\ 0 \end{vmatrix}$

```python
def len_vector(vector: list):
    len_v = 0
    for component in vector:
        len_v += component**2
    return len_v**0.5
```

```python
len_vector([10, 10, 10])
```

2) Пусть задана плоскость: $A_1 \cdot x + B_1 \cdot y + C_1 \cdot z + D1 = 0$
и прямая: 

$\frac{x-x_1}{x_2-x_1} = \frac{y-y_1}{y_2-y_1} = \frac{z-z_1}{z_2-z_1}$

Как узнать, принадлежит прямая плоскости или нет?
Если мы подставим уравнение прямой в уравнение плоскости, то для того, чтобы условие выполнялось необходимо, чтобы выполнялась следующая система уравнений:

$\begin{equation*}
 \begin{cases}
   A_1 \cdot x_1 + B_1 \cdot y_1 + C_1 \cdot z_1 + D1 = 0
   \\
   A_1 \cdot x_2 + B_1 \cdot y_2 + C_1 \cdot z_2 + D1 = 0
 \end{cases}
\end{equation*}$

#### 4. Решите систему уравнений:

$\begin{equation*}
 \begin{cases}
   y = x^2 - 1
   \\
   exp(x) + x \cdot (1 - y) = 1
 \end{cases}
\end{equation*}$

```python
x = np.linspace(-3, 3, 1000)
plt.plot(x, (np.exp(x) - 1 + x)/x)
plt.plot(x, x**2 - 1)
plt.xlabel('x')
plt.ylabel('y')
plt.ylim(-3, 10) 
plt.show()

def eq(p):
    x, y = p
    return (y - x**2 + 1, np.exp(x) + x*(1 - y) - 1)
x1, y1 =  fsolve(eq, (2.5, 6))
print (x1, y1)
```
